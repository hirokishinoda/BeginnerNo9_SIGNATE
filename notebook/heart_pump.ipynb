```python
import pandas as pd
```

```python
import numpy as np
```

```python
train_df = pd.read_csv("../data/train.csv")
```

```python
test_df = pd.read_csv("../data/test.csv")
```

```python
train_df.head()
```

```python
dummies_train = pd.get_dummies(train_df,columns=["anaemia", "diabetes", "high_blood_pressure", "sex", "smoking"],prefix_sep=" ")
dummies_test = pd.get_dummies(test_df,columns=["anaemia", "diabetes", "high_blood_pressure", "sex", "smoking"],prefix_sep=" ")
```

```python
dummies_train["EF_normal"] = 0
dummies_test["EF_normal"] = 0
dummies_train["EF_normal"][dummies_train["ejection_fraction"] >= 50] = 1
dummies_test["EF_normal"][dummies_test["ejection_fraction"] >= 50] = 1
```

```python
dummies_train["EF_abnormal"] = 0
dummies_test["EF_abnormal"] = 0
dummies_train["EF_abnormal"][(dummies_train["ejection_fraction"] < 50) & (dummies_train["ejection_fraction"] >= 40)] = 1
dummies_test["EF_abnormal"][(dummies_test["ejection_fraction"] < 50) & (dummies_test["ejection_fraction"] >= 40)] = 1
```

```python
dummies_train["EF_pseudonormal"] = 0
dummies_test["EF_pseudonormal"] = 0
dummies_train["EF_pseudonormal"][(dummies_train["ejection_fraction"] < 40) & (dummies_train["ejection_fraction"] >= 30)] = 1
dummies_test["EF_pseudonormal"][(dummies_test["ejection_fraction"] < 40) & (dummies_test["ejection_fraction"] >= 30)] = 1
```

```python
dummies_train["EF_restrictive"] = 0
dummies_test["EF_restrictive"] = 0
dummies_train["EF_restrictive"][dummies_train["ejection_fraction"] < 30] = 1
dummies_test["EF_restrictive"][dummies_test["ejection_fraction"] < 30] = 1
```
