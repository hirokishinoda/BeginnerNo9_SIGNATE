```python
import pandas as pd
import lightgbm as lgbm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_data = pd.read_pickle("../src/feature/CPK_test_df.pkl")
train_data = pd.read_pickle("../src/feature/CPK_train_df.pkl")
```

```python
model = lgbm.Booster(model_file="../src/model/lgbm9.txt")
```

```python
feature_imp = pd.DataFrame(model.feature_importance(importance_type='gain'),columns=['importance'])
feature_imp["feature"] = model.feature_name()
feature_imp = feature_imp.sort_values("importance",ascending=False)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='importance',y='feature',data=feature_imp)
plt.tight_layout()
plt.show()
```

```python
drop_feature = ['creatinine_low','sodium_high','creatinine_high','creatinine_normal','EF_normal','platelets_normal','platelets_high','EF_abnormal','CPK_low']
```

```python
train_df = train_data.drop(columns=drop_feature,axis=1)
```

```python
test_df = test_data.drop(columns=drop_feature,axis=1)
```

```python
train_df.to_pickle("../src/feature/select_train_df.pkl")
```

```python
test_df.to_pickle("../src/feature/select_test_df.pkl")
```
