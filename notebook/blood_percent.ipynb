```python
import pandas as pd
```

```python
train_df = pd.read_pickle("../src/feature/select_train_df.pkl")
```

```python
test_df = pd.read_pickle("../src/feature/select_test_df.pkl")
```

```python
train_df.head()
```

```python
train_df["platelets_per_blood"] = train_df["platelets"] / (train_df["platelets"] + train_df["serum_sodium"] + train_df["serum_creatinine"])
test_df["platelets_per_blood"] = test_df["platelets"] / (test_df["platelets"] + test_df["serum_sodium"] + test_df["serum_creatinine"])
```

```python
train_df["sodium_per_blood"] = train_df["serum_sodium"] / (train_df["platelets"] + train_df["serum_sodium"] + train_df["serum_creatinine"])
test_df["sodium_per_blood"] =  test_df["serum_sodium"] / (test_df["platelets"] + test_df["serum_sodium"] + test_df["serum_creatinine"])
```

```python
train_df["creatinine_per_blood"] = train_df["serum_creatinine"] / (train_df["platelets"] + train_df["serum_sodium"] + train_df["serum_creatinine"])
test_df["cretinine_per_blood"] = test_df["serum_creatinine"] / (test_df["platelets"] + test_df["serum_sodium"] + test_df["serum_creatinine"])
```

```python
train_df.to_pickle("../src/feature/blood_train_df.pkl")
```

```python
test_df.to_pickle("../src/feature/blood_test_df.pkl")
```
